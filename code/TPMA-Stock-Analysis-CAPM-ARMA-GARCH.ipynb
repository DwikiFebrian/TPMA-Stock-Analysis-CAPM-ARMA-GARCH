# Hitung Betha1

```r
joinData <- left_join(TPMA %>% select(date, stock_ret = returns), 
                     IHSG %>% select(date, market_ret = returns),
                     by = "date")
```

```r
View(joinData)
```

```r
Cov <- joinData %>% summarise(cov = cov(stock_ret, market_ret, use = "complete.obs")) %>% pull(cov)
VarMark <- joinData %>% summarise(varMark = var(market_ret, na.rm = TRUE)) %>% pull(varMark)

beta1 <- Cov / VarMark
```

```r
hasil <- tibble(
  Stdev = Stdev,
  Cov = Cov,
  VarMark = VarMark,
  beta1 = beta1
)

hasil
```

Nilai beta1: 0.385179530855917

# Expected Return

```r
Rf <- 0.06
Rm <- mean(joinData$market_ret, na.rm = TRUE) * 252

ERi_CAPM <- Rf + beta1 * (Rm - Rf)

ERi_real <- mean(joinData$stock_ret) * 252

tibble(Expected_CAPM = ERi_CAPM, Realized = ERi_real)
```

```r
Sharpe <- (ERi_real - Rf) / sd(joinData$stock_ret)
Sharpe
```

```r
alpha <- ERi_real - ERi_CAPM
alpha
```

Berdasarkan estimasi model CAPM, saham menunjukkan nilai beta dan volatilitas yang rendah, menandakan karakter defensif. Nilai alpha sebesar −0.0113 mengindikasikan bahwa saham tersebut menghasilkan return aktual yang sedikit lebih rendah dari ekspektasi pasar. Nilai Sharpe ratio sebesar −0.42 menegaskan bahwa tingkat pengembalian per unit risiko relatif kurang efisien dibandingkan aset bebas risiko. Hasil ini konsisten dengan karakteristik saham berorientasi dividen yang cenderung stabil, defensif, dan tidak mengejar capital gain tinggi.

# Bangun Model ARMA

```r
ARMA <- arima(TPMA$returns, order = c(2,0,1))
```

```r
print(ARMA)
```

```r
P0 <- TPMA$close[1]
price_reconstructed <- P0 * exp(cumsum(TPMA$returns))
fitted_values <- P0 * exp(cumsum(TPMA$returns - resid(ARMA)))
```

# Plot

```r
ts.plot(TPMA$close)
lines(fitted_values, col=2, lty=2)
```

# Predict

```r
fitted_returns <- TPMA$returns - resid(ARMA)
fitted_values <- P0 * exp(cumsum(fitted_returns))
```

```r
fc <- predict(ARMA, n.ahead=100)
fc_mean <- fc$pred
fc_se   <- fc$se
```

```r
last_price <- tail(price_reconstructed, 1)

forecasted_prices <- last_price * exp(cumsum(fc_mean))
lower <- last_price * exp(cumsum(fc_mean - 2*fc_se))
upper <- last_price * exp(cumsum(fc_mean + 2*fc_se))
```

```r
ts.plot(TPMA$close, xlim=c(1, length(TPMA$close)+100), main="Forecast ARMA → Price")
points((length(TPMA$close)+1):(length(TPMA$close)+100), forecasted_prices, type="l", col=4)
points((length(TPMA$close)+1):(length(TPMA$close)+100), lower, type="l", col=4, lty=2)
points((length(TPMA$close)+1):(length(TPMA$close)+100), upper, type="l", col=4, lty=2)
```

# GARCH

```r
realizedvol <- TPMA %>%
    tq_mutate(
    select     = returns,
    mutate_fun = rollapply,
    width      = 20,
    FUN        = sd.annualized,
    align      = "right",
    col_rename = "realized_vol"
)
```

```r
realizedvol <- drop_na(realizedvol)
```

```r
View(realizedvol)
```

```r
vol <- realizedvol %>% select(date, realized_vol)
```

```r
plot3 <- ggplot(vol, aes(x = date, y = realized_vol))
plot3 + geom_line(color="steelblue")
```

```r
garch_spec <- ugarchspec(variance.model=list(model="eGARCH", garchOrder=c(1,2)),
                         mean.model=list(armaOrder=c(3,1)),
                        distribution.model ="std")
```

```r
fit_garch <- ugarchfit(spec = garch_spec, data = vol$realized_vol)
```

```r
fit_garch
```

# Diagnosis

```r
resid_std <- residuals(fit_garch, standardize=TRUE)

ts.plot(resid_std, main="Standardized Residuals", type="l")
hist(resid_std, breaks=30, main="Histogram of Standardized Residuals")
```

```r
# Ljung-Box test for standardized residuals
Box.test(resid_std, lag=30, type="Ljung-Box")

# Ljung-Box test for squared standardized residuals
Box.test(resid_std^2, lag=30, type="Ljung-Box")
```

```r
qqnorm(resid_std)
qqline(resid_std, col="red")
```

# Forecast

```r
forecast <- ugarchforecast(fit_garch, n.ahead = 20)

sigma_forecast <- sigma(forecast) 
mu_forecast <- fitted(forecast)
```

```r
sigma_forecast
```

```r
mu_forecast
```

# VaR

```r
alpha <- c(0.01, 0.05)  # level VaR

nu <- coef(fit_garch)["shape"]

q_t <- qt(alpha, df = nu)

VaR <- sapply(alpha, function(a) mu_forecast + sigma_forecast * qt(a, df = nu))
VaR
```

```r
last_date <- last(vol$date)
dates_forecast <- seq(from = last_date + 1, by = 1, length.out = 20)
```

```r
sigma_xts <- xts(sigma_forecast, order.by = dates_forecast)
VaR_xts <- xts(VaR, order.by = dates_forecast)
```

```r
plot(sigma_xts, main="Forecast Volatility", type="l")
matplot(dates_forecast, VaR, type="l", lty=1:2, col=c("red","blue"), main="Forecast VaR 1% & 5%")
legend("topleft", legend=c("VaR 1%","VaR 5%"), col=c("red","blue"), lty=1:2)
```
